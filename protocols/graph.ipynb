```python
import matplotlib.pyplot as plt
import numpy as np

# Data for visualizations (assuming hypothetical results based on protocols' performance)
protocols = ['BB84', 'B92', 'E91', 'SSP']

# Example data for key similarity percentage after eavesdropping
key_similarity = [85, 80, 95, 75]  # Percentages of similarity after potential eavesdropping

# Example data for error rates (in percentage)
error_rates = [2, 3, 1, 4]  # Hypothetical error rates for the protocols

# Example data for CHSH parameter values (E91 specific, range: 0 to 22 ~ 2.828)
chsh_values = [None, None, 2.6, None]  # Only E91 has CHSH values

# Visualization for Key Similarity Percentage
plt.figure(figsize=(10, 6))
plt.bar(protocols, key_similarity, color='skyblue')
plt.xlabel('Protocols')
plt.ylabel('Key Similarity Percentage (%)')
plt.title('Key Similarity Percentage After Eavesdropping for QKD Protocols')
plt.show()

# Visualization for Error Rates
plt.figure(figsize=(10, 6))
plt.bar(protocols, error_rates, color='salmon')
plt.xlabel('Protocols')
plt.ylabel('Error Rates (%)')
plt.title('Error Rates for QKD Protocols')
plt.show()

# Visualization for CHSH Values (only for E91 protocol)
if chsh_values[2] is not None:
    plt.figure(figsize=(6, 6))
    plt.bar(['E91'], [chsh_values[2]], color='limegreen')
    plt.xlabel('Protocol')
    plt.ylabel('CHSH Parameter')
    plt.title('CHSH Parameter Value for E91 Protocol')
    plt.ylim(0, 2.828)  # Set range up to the maximum CHSH value
    plt.show()
```
